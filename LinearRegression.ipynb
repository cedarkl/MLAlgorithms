```python
data = pd.read_csv(open("iris.csv"))
data.head()
```

>From **petal_width** => **petal_length**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.regplot(data=data, x='petal_width', y='petal_length')
plt.show()
```

```python
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
input_ = data[['petal_width']]
output_ = data[['petal_length']]
```

```python
regression_model = linear_model.LinearRegression()
regression_model = regression_model.fit(input_, output_)
prediction = regression_model.predict([[1.5]])
print(prediction)
```
