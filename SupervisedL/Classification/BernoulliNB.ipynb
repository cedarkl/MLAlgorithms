```python
import pandas as pd
```

```python
# loading the data from csv file to a Panda Dataframe
data = pd.read_excel('playgolf_data.xlsx', index_col=0)

# printing the first 5 rows of the dataframe
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
features = data.drop("Play Golf", axis=1)
target = data[['Play Golf']]
```

```python
features.columns
```

```python
# features = pd.get_dummies(features, dtype=int, drop_first=True)
```

```python
features = pd.get_dummies(features, dtype=int)
features
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.validation import column_or_1d
le = LabelEncoder()
target = le.fit_transform(column_or_1d(target))
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(features, column_or_1d(target))
```

```python
classifier.score(features, target)
```

```python
from sklearn.naive_bayes import BernoulliNB
classifier = BernoulliNB()
classifier.fit(features, column_or_1d(target))
```

```python
classifier.score(features, target)
```

```python
class_names = classifier.classes_
class_names
```

```python
import pickle
pkl_file = "pickle_classifier.pkl"
with open(pkl_file, 'wb') as file:
    pickle.dump(classifier, file)
```

```python
with open(pkl_file, 'rb') as file:
    pickle_classifier = pickle.load(file)
```

```python
X_check = [[1, 0, 0, 1, 0, 0, 1, 0, 1, 0],
           [0, 0, 1, 1, 0, 0, 1, 0, 0, 1]]
y_result = pickle_classifier.predict(X_check)
y_result
```
