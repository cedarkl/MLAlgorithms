```python
import pandas as pd
```

```python
# loading the data from csv file to a Panda Dataframe
data = pd.read_excel('evade.xlsx', index_col=0)

# printing the first 5 rows of the dataframe
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
features = data[['Refund', 'Marital_Status', 'Taxable_Income']]
target = data[['Evade']]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.utils.validation import column_or_1d
classifier = GaussianNB()
classifier.fit(features, column_or_1d(target))
```

```python
import numpy as np
classifier.score(features, np.array(target))
```

```python
class_names = classifier.classes_
class_names
```

```python
# X_check = [["No", "Married", 120000],
#            ["No", "Single", 90000]]
X_check = [[0, 1, 120000],
           [0, 0, 90000]]
y_result = classifier.predict(X_check)
y_result
```

```python
import pickle
pkl_file = "pickle_classifier.pkl"
with open(pkl_file, 'wb') as file:
    pickle.dump(classifier, file)
```

```python
with open(pkl_file, 'rb') as file:
    pickle_classifier = pickle.load(file)
```

```python
x_check = [[0, 0, 75000]]
y_result_ = pickle_classifier.predict(x_check)
y_result_
```
