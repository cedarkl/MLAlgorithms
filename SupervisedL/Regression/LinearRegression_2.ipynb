```python
import pandas as pd
data = pd.read_csv(open('iris.csv'))
```

>From **petal_length**, **petal_width**, **petal_length**  => **petal_width**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(data)
plt.show()
```

```python
data.columns
```

```python
inputs = data[['sepal_length', 'sepal_width', 'petal_length']]
output_ = data[['petal_width']]
```

```python
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(inputs, output_, test_size=0.2)

regression_model = linear_model.LinearRegression()
regression_model = regression_model.fit(X_train, y_train)
y_pred = regression_model.predict(X_valid)
```

```python
m = regression_model.coef_[0]
b = regression_model.intercept_
print("Coefficients= ", m)
print("Intercept= ", b)
```

```python
print("mean squared error= %.2f" % mean_squared_error(output_, regression_model.predict(inputs)))
print("variance score= %.2f" % regression_model.score(inputs, output_))
```

```python
#check the score of train and valid data
regression_model.score(X_valid, y_valid)
```
